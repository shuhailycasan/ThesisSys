```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
con = pd.read_csv('cor3.csv')
con
```

```python
con.corr()
```

```python
hm = sns.heatmap(con.corr(method= 'spearman'), annot = True)

hm.set(xlabel='\nDetails', ylabel='Details\t', title = "Correlation matrix of Senior High and College Students \n")

plt.show()
```

```python
plt.savefig("Plotting_Correlation_HeatMap_With_Axis_Titles.png")
```

```python
sns.regplot(x=con["answe1"], y=con["answer3"])

plt.savefig("Plotting_Correlation_Scatterplot_With_Regression_Fit.jpg")
```
